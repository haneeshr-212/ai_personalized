```python
import json

# Replace with your actual uploaded file name
html_filename = 'index.html'
output_notebook = 'converted_cells.ipynb'

# Read the HTML content
with open(html_filename, 'r', encoding='utf-8') as f:
    html_content = f.read()

# Define the basic .ipynb structure
# Every line of HTML content is put into a code cell source list
notebook_data = {
    "cells": [
        {
            "cell_type": "code",
            "execution_count": None,
            "metadata": {},
            "outputs": [],
            "source": [html_content] # Content goes here
        }
    ],
    "metadata": {
        "kernelspec": {
            "display_name": "Python 3",
            "language": "python",
            "name": "python3"
        }
    },
    "nbformat": 4,
    "nbformat_minor": 0
}

# Save the new notebook file
with open(output_notebook, 'w', encoding='utf-8') as f:
    json.dump(notebook_data, f)

print(f"Successfully created {output_notebook}")
```
